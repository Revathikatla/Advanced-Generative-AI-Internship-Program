# **1.Smart Parking Lot Management System**


```python
def parking(capacity, logs):
  count = 0
  peak = 0
  for i in logs:
    if i == "IN":
      count += 1
    elif i == "OUT" and count > 0:
      count -= 1
      if count > peak:
        peak = count
  print("Currently Parked Vehicles:", count)
  print("Peak Parking Usage:", peak)
  if count > capacity:
    print("Parking Status: Capacity Exceeded")
  else:
    print("Parking Status: Available")
parking(50, ["IN","IN","IN","OUT","IN","IN","OUT"])
```

# **2.Online Food Delivery Time Estimator**


```python
def delivery_time(distance, traffic, weather):
  time = distance * 5
  if traffic == "High":
    time += 15
  elif traffic == "Medium":
    time += 10
  else:
    time += 5
  if weather == "Rainy":
    time += 10
  elif weather == "Storm":
    time += 20
  print("Estimated Delivery Time:", time, "minutes")
delivery_time(8, "High", "Rainy")
```

# **3.Movie Theatre Seat Occupancy Analyzer**


```python
def seat_analysis(total_seats, booked_seats):
  booked = len(booked_seats)
  percent = (booked / total_seats) * 100
  print("Occupancy", int(percent), "%")
  if percent == 100:
    print("Show Status: Housefull")
    print("Suggestion: Open Additional Show")
  elif percent >= 75:
    print("Show Status: Almost Full")
  else:
    print("Show Status: Seats Available")
seat_analysis(200, [1]*150)
```

# **4.Cloud Server Load Classification System**

```python
def server_load(cpu_readings):
  total = 0
  for i in cpu_readings:
    total += i
  avg = total / len(cpu_readings)
  print("Average CPU Load:", int(avg), "%")
  if avg < 50:
    print("Server Status: Normal")
  elif avg <= 80:
    print("Server Status: Warning")
  else:
    print("Server Status: Critical")
server_load([45, 60, 70, 85, 90])
```

# **5.Smart Classroom Resource Usage Monitor**


```python
def resource_check(usage):
    overused = []
    for resource, hours in usage.items():
        if hours > 8:      # limit set as 8 hours
            overused.append(resource)
    if overused:
        print("Overused Resources:", ", ".join(overused))
        print("Energy Alert: Yes")
    else:
        print("Overused Resources: None")
        print("Energy Alert: No")
usage_data = {
    "Projector": 6,
    "AC": 9,
    "Lights": 4
}
resource_check(usage_data)
```

# **6. Online Event Registration Capacity Controller**


```python
def event_registration(capacity, registrations):
  confirmed = 0
  waitlist = 0
  for i in range(registrations):
    if confirmed < capacity:
      confirmed += 1
    else:
      waitlist += 1
  print("Confirmed Registrations:", confirmed)
  print("Waitlisted Users:", waitlist)
  if waitlist > 0:
    print("Registration Status: Closed")
  else:
    print("Registration Status: Open")
event_registration(100, 105)
```
