# **1. Smart Light Controller**


```python
class Light:
  def __init__(self, name):
    self.name = name
    self.status = "OFF"
  def turn_on(self):
    self.status = "ON"
  def show(self):
    print(self.name, "is", self.status)
l = Light("Bedroom Light")
l.turn_on()
l.show()
```

# **2. Employee ID Card System**


```python
class Employee:
  def __init__(self, name, emp_id, dept):
    self.name = name
    self.emp_id = emp_id
    self.dept = dept
  def show(self):
    print("Name:", self.name)
    print("Id:", self.emp_id)
    print("Department:", self.dept)
e = Employee("Rahul", "EMP102", "IT")
e.show()
```

# **3. Mobile Contact Record**


```python
class Contact:
  def __init__(self, name, phone):
    self.name = name
    self.phone = phone
  def show(self):
    print("Contact Saved")
    print("Name:", self.name)
    print("Phone:", self.phone)
c = Contact("Shailu", "9876543210")
c.show()
```

# **4. Product Price Tag Generator**


```python
class Product:
  def __init__(self, name, price):
    self.name = name
    self.price = price
  def show(self):
    print("Product:", self.name)
    print("Price:" + str(self.price))
p = Product("Headphones", 3000)
p.show()
```

# **5. Movie Rating Display System**


```python
class Movie:
    def __init__(self, name, rating):
        self.name = name
        self.rating = rating
    def show(self):
        print("Movie:", self.name)
        print("Rating:", self.rating)
m = Movie("Inception", 4.8)
m.show()
```

# **6. Delivery Address Manager**


```python
class Delivery:
  def __init__(self, name, address):
    self.name = name
    self.address = address
  def show(self):
    print("Delivery Details")
    print("Customer:", self.name)
    print("Address:", self.address)
d = Delivery("Shailu", "Karimnagar")
d.show()
```
