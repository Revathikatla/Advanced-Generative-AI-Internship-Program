## 1.User Login Check

## 2.Pass/Fail Analyzer

```python
marks = [45, 78, 90, 33, 60]
num_pass = 0
for i in marks:
  if i >= 50:
    num_pass = num_pass + 1
fail = len(marks) - num_pass
print("Pass:", num_pass)
print("Fail:", fail)
```

## 3.Simple Data Cleaner

```python
names = [" Alice ", "bob", " CHARLIE "]
cleaned_names = []
for name in names:
  cleaned_name = name.strip().lower()
  cleaned_names.append(cleaned_name)
print(cleaned_names)
```

## 4.Message Length Analyzer

```python
messages = ["Hi", "Welcome to the platform", "OK"]

for message in messages:
  length = len(message)
  print(f"Message: '{message}' (Length: {length})")
  if length > 10:
    print(" This message is longer than 10 characters!")
```

## 5.Error Message Detector

```python
logs = ["INFO", "ERROR", "WARNING", "ERROR"]
error_count = 0
for entries in logs:
  if entries == "ERROR":
    error_count += 1
print(f"Number of ERROR messages: {error_count}")
```
