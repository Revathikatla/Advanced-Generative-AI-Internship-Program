## Problem 1: Unique Words in a Sentence


```python
sentence = input("Enter sentence: ")

# Split the sentence into words
words = sentence.lower().split()

# Convert the list of words into a set to remove duplicates
unique_words = set(words)

# Print the count of unique words and the unique words themselves
print(f"Unique words count: {len(unique_words)}")
print(f"Unique words: {unique_words}")
```

## Problem 2: Highest Salary from Employee Data


```python
employees = {
    "Ravi": 75000,
    "Anita": 68000,
    "Kiran": 72000,
    "Priya": 80000,
    "Mohan": 65000
}

# Initialize variables to store the highest salary and the employee with that salary
highest_salary = 0
highest_paid_employee = ""

# Iterate through the dictionary to find the highest salary
for name, salary in employees.items():
    if salary > highest_salary:
        highest_salary = salary
        highest_paid_employee = name

# Print the result in the specified format
print(f"Highest Salary: {highest_paid_employee} - {highest_salary}")
```

## Problem 3: Find Maximum and Minimum Values


```python
numbers = [45, 22, 89, 10, 66]

# Initialize max and min with the first element of the list
maximum = numbers[0]
minimum = numbers[0]

# Iterate through the list starting from the second element
for num in numbers[1:]:
    if num > maximum:
        maximum = num
    if num < minimum:
        minimum = num

# Print the results in the specified format
print(f"List: {numbers}")
print(f"Max: {maximum}")
print(f"Min: {minimum}")
```

## Problem 4: Count Products Above a Price Threshold

```python
prices = [450, 1200, 899, 1500, 300]

# Initialize a counter for products above the threshold
count_above_threshold = 0
threshold = 1000

# Iterate through the list of prices
for price in prices:
    if price > threshold:
        count_above_threshold += 1

# Print the result in the specified format
print(f"Products above {threshold}: {count_above_threshold}")
```

## Problem 5: Calculate Attendance Percentage

```python
attendance = ["P", "P", "A", "P", "P"]

# Count the number of 'P' (Present) entries
present_count = attendance.count("P")

# Get the total number of days (length of the list)
total_days = len(attendance)

# Calculate the attendance percentage
attendance_percentage = (present_count / total_days) * 100

# Print the result in the specified format
print(f"Attendance Percentage: {attendance_percentage:.1f}")
```
